```python
# 1. Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

```python
# 2. Load Dataset
df = pd.read_csv('../data/raw/data.csv')  # update filename if different
print("Dataset shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
# 3. Data Overview
df.info()
df.head()
```

```python

# 4. Summary Statistics
df.describe(include='all').T
```

```python
# 5. Distribution of Numerical Features
num_cols = df.select_dtypes(include=np.number).columns

df[num_cols].hist(bins=30, figsize=(15, 10), edgecolor='black')
plt.suptitle("Distributions of Numerical Features", fontsize=16)
plt.tight_layout()
plt.show()

```

```python
# 6 Select categorical features to visualize
selected_cats = ['ProductCategory', 'ChannelId', 'CurrencyCode', 'ProviderId', 'ProductId']

for col in selected_cats:
    plt.figure(figsize=(10, 4))
    
    if df[col].nunique() > 15:
        # Only plot top 10 values if too many categories
        top_10 = df[col].value_counts().nlargest(10)
        sns.barplot(x=top_10.index, y=top_10.values)
        plt.title(f"Top 10 Most Frequent Values in {col}")
    else:
        sns.countplot(data=df, x=col, order=df[col].value_counts().index)
        plt.title(f"Distribution of {col}")
    
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# 7. Correlation Matrix (Numerical Features)
plt.figure(figsize=(12, 8))
corr_matrix = df[num_cols].corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation Matrix of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
# 8. Missing Values Analysis
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

if not missing.empty:
    print("Missing Values Summary:")
    print(missing)
else:
    print("No missing values detected.")
```

```python
# 9. Outlier Detection via Boxplots
for col in num_cols:
    plt.figure(figsize=(6, 3))
    sns.boxplot(data=df, x=col)
    plt.title(f"Boxplot for {col} (Outlier Detection)")
    plt.tight_layout()
    plt.show()
```
